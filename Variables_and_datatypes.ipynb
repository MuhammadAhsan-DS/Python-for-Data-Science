### `Rules for giving names (identifiers) in Python`


````python

# 1) Use only letters (A-Z, a-z), digits (0-9), and underscore (_).
# 2) A name must start with a letter or underscore, not with a digit.
# 3) Spaces and special symbols are not allowed: @, #, $, %, -, +, etc.
# 4) Python keywords cannot be used as names (for, if, class, True, None, ...).
# 5) Names are case-sensitive (age, Age, and AGE are different).
# 6) Use meaningful names to make code readable (e.g., total_marks, user_name).
# 7) Recommended style (PEP 8):
#    - Variables/functions: snake_case
#    - Constants: UPPER_CASE
#    - Classes: PascalCase

# Valid examples:
student_name = "Ahsan"
_total = 100
age2 = 20
MAX_RETRIES = 3

# Invalid examples (kept as strings so the cell runs without syntax errors):
invalid_names = [
    "2name",      # starts with a digit
    "user-name",  # contains hyphen
    "first name", # contains space
    "class",      # keyword
    "@score"      # special character
]

print("Python naming rules added with valid and invalid examples.")
````

### `Data Types in python`

````python
v = 2+3j
print("Complex number:", v)
print("Data type of complex number:", type(v))
````

````python
num = [2,3,4,5,7]
print("List of numbers:", num)
print("Data type of num:", type(num))
````

````python
a = True
print("Boolean value:", a)
print("Data type of boolean value:", type(a))
````

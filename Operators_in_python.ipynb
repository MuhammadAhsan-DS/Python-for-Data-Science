# `Python Operators for Data Science`

Operators are symbols (or keywords) that tell Python to perform operations on values and variables.
In data science, operators are the foundation of data cleaning, feature engineering, filtering, comparison, and logical decision-making.

## `Why Operators Matter in Data Science`

- Build **features**: create ratios, differences, growth rates, normalized values.
- Filter data: select rows using conditions (for example, age > 30 and city == "Lahore").
- Compare models: check performance metrics (accuracy_a > accuracy_b).
- Handle missing data and defaults: use membership and logical checks.
- Work with arrays and DataFrames: perform vectorized operations efficiently.

### `1. Arithmetic Operators`

Used for mathematical computation.

| Operator | Meaning | Example | Data Science Use Case |
|---|---|---|---|
| `+` | Addition | `a + b` | Combine signals, aggregate features |
| `-` | Subtraction | `a - b` | Compute residuals/errors |
| `*` | Multiplication | `a * b` | Scaling, weighted features |
| `/` | Division | `a / b` | Ratios, rates, normalization |
| `//` | Floor division | `a // b` | Bucketing into integer groups |
| `%` | Modulus | `a % b` | Cyclical features (hour % 24) |
| `**` | Power | `a ** b` | Polynomial feature creation |

```python
# `Example: feature engineering`
df["income_per_person"] = df["household_income"] / df["household_size"]
df["is_weekend"] = df["day_of_week"] % 7 >= 5
```



````python
# A grocery store sold 125 apples at Rs. 30 each and gave a flat discount of Rs. 450.
# Write a Python expression using arithmetic operators to find the final revenue.

Apple_sold = 125
Price_per_apple = 30
Discount = 450
Revenue = (Apple_sold * Price_per_apple) - Discount
print("The final revenue is:", Revenue)
````

````python
# An e-commerce store sold 480 headphones at Rs. 2,750 each; 12% of units were returned, and each return incurred a reverse-logistics cost of Rs. 180.
# If the store also spent Rs. 95,000 on ads and then paid 8% GST on net sales (after returns), write one Python expression using arithmetic operators to compute final profit.
Headphones_sold = 480
Price_per_headphone = 2750
Returns_percentage = 0.12
Returns_cost_per_unit = 180
Ads_cost = 95000
Units_returned = Headphones_sold * Returns_percentage
Net_sales = (Headphones_sold - Units_returned) * Price_per_headphone
GST = 0.08 * Net_sales
Profit = Net_sales - (Units_returned * Returns_cost_per_unit) - Ads_cost - GST
print("The final profit is:", Profit)
````

### `2. Comparison (Relational) Operators`

Return boolean values: `True` or `False`.

| Operator | Meaning |
|---|---|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater than or equal to |
| `<=` | Less than or equal to |

```python
# Example: filtering high-value customers
high_value = df[df["total_spend"] > 10000]
```


````python
# A telecom company predicts customer churn probability for each user; write a condition using comparison operators to flag a user as high-risk when churn_probability >= 0.70 and monthly_usage < 120.
# Then check whether the predicted probability is exactly equal to the observed churn label threshold (0 or 1) using an appropriate comparison operator.
churn_probability = 0.75
monthly_usage = 100
is_high_risk = (churn_probability >= 0.70) and (monthly_usage < 120)
print("Is the user high-risk?", is_high_risk)
# Then check whether the predicted probability is exactly equal to the observed churn label threshold (0 or 1) using an appropriate comparison operator.
observed_churn_label = 1
is_churn_label_match = (churn_probability == observed_churn_label)
print("Does the predicted probability match the observed churn label?", is_churn_label_match)
````

### `3. Logical Operators`

Combine multiple conditions.

| Operator | Meaning | Example |
|---|---|---|
| `and` | True if both are true | `a > 0 and b > 0` |
| `or` | True if at least one is true | `score > 90 or attendance > 95` |
| `not` | Reverses truth value | `not is_missing` |

```python
# Pure Python logical checks
if (accuracy > 0.9) and (f1_score > 0.85):
    print("Model is production-ready candidate")
```

Note: In pandas, use `&`, `|`, and `~` (with parentheses) for element-wise logic.

```python
# Pandas boolean masking
subset = df[(df["age"] >= 18) & (df["city"] == "Karachi")]
```



````python
# A hospital wants to prioritize patients for immediate review: mark a patient as "critical" if (heart_rate > 120 and oxygen_level < 90) or not has_recent_checkup.
# Write a Python logical condition using and, or, and not to implement this triage rule.
heart_rate = 130
oxygen_level = 85
has_recent_checkup = False
is_critical = (heart_rate > 120 and oxygen_level < 90) or not has_recent_checkup
print("Is the patient critical?", is_critical)
````

### `4. Assignment Operators`

Assign or update variable values.

| Operator | Meaning | Example |
|---|---|---|
| `=` | Assign | `x = 10` |
| `+=` | Add and assign | `x += 2` |
| `-=` | Subtract and assign | `x -= 1` |
| `*=` | Multiply and assign | `x *= 3` |
| `/=` | Divide and assign | `x /= 2` |
| `//=` | Floor divide and assign | `x //= 2` |
| `%=` | Modulus and assign | `x %= 5` |
| `**=` | Power and assign | `x **= 2` |

```python
# `Example: iterative metric update`
running_total = 0
for value in batch_values:
    running_total += value
```



````python
# A food-delivery platform tracks daily revenue in a variable `revenue`; over three updates, add Rs. 12,500, subtract Rs. 2,300 refund losses, then apply a 5% campaign boost using assignment operators only.
# Write the Python statements using +=, -=, and *= to update `revenue` step by step.
revenue = 0
revenue += 12500  # Add daily revenue
revenue -= 2300   # Subtract refund losses
revenue *= 1.05   # Apply 5% campaign boost
print("The final revenue after updates is:", revenue)
````

### `5. Membership Operators`

Check if a value exists in a sequence/container.

| Operator | Meaning |
|---|---|
| `in` | Present in container |
| `not in` | Not present in container |

```python
allowed_cities = {"Lahore", "Karachi", "Islamabad"}
df = df[df["city"].isin(allowed_cities)]
```



````python
# A fraud-detection model should flag transactions when the payment country is in a high-risk watchlist and ignore transactions when the merchant category is not in approved_categories.
# Write Python conditions using `in` and `not in` to implement these two checks.
payment_country = "Nigeria"

high_risk_countries = ["North Korea", "Iran", "Nigeria"]

merchant_category = "Cryptocurrency"

approved_categories = ["Groceries", "Electronics", "Clothing"]

is_flagged = payment_country in high_risk_countries
is_ignored = merchant_category not in approved_categories

print("Is the transaction flagged?", is_flagged)
````

### `6. Identity Operators`

Check if two variables refer to the same object in memory.

| Operator | Meaning |
|---|---|
| `is` | Same object |
| `is not` | Different objects |

```python
# Common data science pattern
if df is not None:
    print("DataFrame loaded")
```

Use `is` mainly for `None` checks, not for numeric/string equality.



````python
# A data pipeline should run model training only if the loaded DataFrame object is not None; otherwise it should stop and log an error.
# Write a Python condition using identity operators (`is` / `is not`) to perform this safety check.
data = None  # Simulating a failed data load
if data is not None:
    print("Data loaded successfully. Proceeding with model training.")
else:
    print("Error: Data loading failed. Model training cannot proceed.")
````

### `7. Bitwise Operators`

Operate at binary (bit) level.

| Operator | Meaning |
|---|---|
| `&` | Bitwise AND |
| `|` | Bitwise OR |
| `^` | Bitwise XOR |
| `~` | Bitwise NOT |
| `<<` | Left shift |
| `>>` | Right shift |

In data science, these are less common for raw analytics but useful in:
- Binary flags and compact encoding
- Low-level optimization tasks
- Some image/signal processing operations


````python
# A smart-city sensor system stores alert states in one integer bitmask (bit 0 = traffic jam, bit 1 = air pollution, bit 2 = flood risk).
# Write Python bitwise conditions to check whether traffic and pollution are both active using `&`, and then turn on flood risk using `|`.
alert_state = 0b001  # Only traffic jam is active
# Check if both traffic and pollution are active
traffic_and_pollution_active = (alert_state & 0b011) == 0b011
print("Are both traffic and pollution active?", traffic_and_pollution_active)
# Turn on flood risk
alert_state = alert_state | 0b100
print("Updated alert state:", bin(alert_state))
````


### `Operator Precedence (Important)`

Python evaluates some operators before others. Use parentheses to avoid mistakes.

```python
# Safer and clearer
mask = (df["sales"] > 5000) & (df["region"] == "South")
```

### `Quick Data Science Cheat Sheet`

- Use arithmetic operators for feature creation.
- Use comparison + logical operators for filtering and segmentation.
- Use assignment operators for iterative calculations.
- Use membership operators for category validation.
- Use identity operator (`is None`) for null-object checks.
- Use parentheses generously for readable, bug-free boolean logic.
